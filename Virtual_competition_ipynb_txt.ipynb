**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df=pd.DataFrame(data,index=labels)
df
````

Diabetes_model


````python
df=pd.read_csv('/content/diabetes_model.csv')
df
````

````python
import seaborn as sns
````

````python
#histogram

plt.hist(df['BloodPressure'],bins=10,color='red',edgecolor='black')
plt.xlabel('BloodPressure')
plt.ylabel('Range')
plt.title('Histogram of BloodPressure')
plt.show()
````

The histogram displays the distribution of the 'BloodPressure' variable from the dataset. With 10 bins, this visualization shows how blood pressure values are spread across different ranges. The red bars with black edges highlight the frequency of values within each range. This plot helps to identify the most common blood pressure ranges and assess the overall distribution, revealing patterns or potential outliers in the dataset.

````python
#histogram of BMi

plt.hist(df['BMI'],bins=10,color='pink',edgecolor='black')
plt.xlabel('BMI')
plt.ylabel('Range')
plt.title('Histogram of BMI')
plt.show()
````


The histogram visualizes the distribution of the 'BMI' values from the dataset. By dividing the data into 10 bins, the plot illustrates how BMI values are distributed across different ranges. The pink bars with black edges make it easy to see the frequency of BMI values within each range. This histogram helps in understanding the overall spread of BMI measurements and can highlight common BMI ranges as well as any significant deviations or outliers.

````python
#bivariate Analysis

#scatter plot
sns.scatterplot(x='Age',y='BMI',data=df)
plt.title('Scatter Plot of Age and BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()

````

The scatter plot illustrates the relationship between 'Age' and 'BMI' in the dataset. Each point represents an individual, with the x-axis showing age and the y-axis showing BMI. This visualization helps identify any patterns or correlations between age and BMI, such as whether BMI tends to increase or decrease with age,

````python
#correlation matrices
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.show()
````

multi analysis


````python
# pairplot

sns.pairplot(df[['Glucose']])
plt.show()
````

The pairplot of the 'Glucose' variable, while displaying only one feature, provides a visual representation of its distribution and relationships. Since there is only one variable plotted, the pairplot will effectively show a single histogram of 'Glucose' values. This helps in understanding the distribution and frequency of glucose levels in the dataset, highlighting any skewness or concentration of values.
